```python
import numpy as np
import pandas as pd
import os
import time
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras import layers 
from tensorflow.keras.preprocessing import image
from tensorflow.keras.layers.experimental import preprocessing
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
```

## Fetching Directories
Mengonfirmasi total data sample yang terdapat pada dataset. Dataset yang sebenarnya berjumlah 2188 data sample dan memiliki 3 kelas

```python
# Timing Starts here
start_time = time.time()

path_directory = '/content/rockpaperscissors/'
print(os.listdir(path_directory))

# Total sample confirmation (1314 + 874)
total_paper = len(os.listdir(os.path.join(path_directory, 'rps-cv-images/paper')))
total_rock = len(os.listdir(os.path.join(path_directory, 'rps-cv-images/rock')))
total_scissors = len(os.listdir(os.path.join(path_directory, 'rps-cv-images/scissors')))
total_dataset = total_paper + total_rock + total_scissors
print("Total Sample : "+str(total_dataset))
```

## Data Augmentation

Diganti dari yang sebelumnya menggunakan fungsi Sequential untuk data augmentation menjadi menggunakan fungsi ImageDataGenerator(). mengikuti persyaratan. 

Set data validation splitting 40% dilakukan pada fungsi ImageDataGenerator() dan menginisialisasi masing2 fungsi dengan attribute "subset" pada pipeline(cell berikutnya) untuk menandakan yang mana training data yang mana validation data

Saran: Berdasarkan dokumentasi tensorflow untuk fungsi ImageDataGenerator() kurang direkomendasikan pada tautan berikut (https://www.tensorflow.org/api_docs/python/tf/keras/preprocessing/image/ImageDataGenerator) dan disarankan menggunakan cara saya yang sebelumnya yaitu dengan fungsi image_from_directory() dan data augmentasi langsung pada model fungsi Sequential(). Secara konsep sama namun beda penulisan saja


```python
img_height = 150
img_width = 150
batch_size = 32

train_datagen = image.ImageDataGenerator(
    validation_split=0.4,
    rescale=1./255,
    horizontal_flip=True,
    vertical_flip=True,
    rotation_range=20,
    zoom_range=0.2,
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    shear_range=0.1,
    fill_mode="nearest"
)
test_datagen = image.ImageDataGenerator(
    validation_split=0.4,
    rescale=1./255
)
```

## Data Pipeline + Train Test Split
**Input Shape** => [150, 150] <br>
**Label Mode** => Categorical | [1, 0 ,0], [0, 1, 0], [0, 0, 1] 

```python
train_generator = train_datagen.flow_from_directory(
    path_directory+'rps-cv-images',
    class_mode='categorical',
    subset='training', 
    seed=42,
    target_size=(150, 150),
    batch_size=32
)
validation_generator = test_datagen.flow_from_directory(
    path_directory+'rps-cv-images',
    class_mode='categorical',
    subset='validation', 
    seed=42,
    target_size=(150, 150),
    batch_size=32
)

class_names = np.unique(train_generator.labels)
```

## Visualizing Dataset Images

```python
plt.figure(figsize = (15, 30))

for enum, images in zip(range(30), train_generator):
    plt.subplot(6, 5, enum+1)
    plt.subplots_adjust(hspace=0.1, wspace=0.3)
    plt.imshow(images[0][enum])
#     plt.title(f'labels: {images[1][enum][0]}')
    if images[1][enum][0] == 1:
        plt.title('labels: paper')
    elif images[1][enum][1] == 1:
        plt.title('labels: rock')
    elif images[1][enum][2] == 1:
        plt.title('labels: scissors')
    else:
        plt.title('labels: N/A')
```

## CNN Model

```python
model = Sequential([
    layers.Conv2D(32, (3,3), input_shape=(150, 150, 3), activation='relu'),
    layers.MaxPooling2D(2,2),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),
    layers.Flatten(),
    layers.Dense(512, activation='relu'),
    layers.Dense(len(class_names), activation='softmax'),
])
```

```python
model.compile(loss='categorical_crossentropy', optimizer=tf.keras.optimizers.Adam(lr = 0.0001), metrics=['accuracy'])
```

```python
model.summary()
```

## Define Callbacks
- Model Checkpoints => For saving the best model monitored metric there is for future evaluation metrics
- Early Stopping => Stop training when monitored metric has stopped improving

```python
filepath = "checkpoints.h5"
accuracy_history = ModelCheckpoint(
    filepath, 
    monitor='val_accuracy', 
    verbose=1, 
    save_best_only=True, 
    mode='max', 
    save_weights_only=False
)

early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=5, 
    min_delta=0,
    verbose=1,
    restore_best_weights=True
)

callbacks_list = [
    accuracy_history,
    early_stopping
]
```

```python
training_history = model.fit(
    train_generator,
    validation_data=(validation_generator),
    epochs=50,
    callbacks=callbacks_list,
    verbose=1,
)
```

## Accuracy & Training Loss Graphics

```python
plt.title("Accuracy")
plt.plot(training_history.history['accuracy'], label='accuracy')
plt.plot(training_history.history['val_accuracy'], label='validation accuracy')
plt.legend()
```

```python
plt.title("Training Loss")
plt.plot(training_history.history['loss'], label='training loss')
plt.plot(training_history.history['val_loss'], label='validation loss')
plt.legend()
```

## Fetching All Validation Samples for Evaluation Metrics
Pemanggilan data validation kembali menggunakan flow_from_directory() disini dikarenakan saya masih kurang paham dengan penerapan batch_size jika ingin mengambil semua sample validation dari fungsi pemanggilan data validation sebelumnya yang telah di set dengan 32 batch_size. Hal ini diperlukan agar total evaluation metrics dapat mengevaluasi seluruh (874 data) sample dibanding hanya mengevaluasi satu batch (32 data) saja.  

```python
test_datagen = image.ImageDataGenerator(
    validation_split=0.4,
    rescale=1./255
)
validation_generator = test_datagen.flow_from_directory(
    path_directory+'rps-cv-images',
    class_mode='categorical',
    subset='validation', 
    seed=42,
    target_size=(150, 150),
    batch_size=874
)

X_valid, y_valid = validation_generator[0][0], validation_generator[0][1]

model_saved = tf.keras.models.load_model('checkpoints.h5')
y_pred = model_saved.predict(X_valid)
pred = np.argmax(y_pred,axis=1)
print(pred)
```

## Evaluation Metrics Classification Report

Expected results (874 data samples):
- Accuracy >= 98%

```python
pred_labels = ["Paper", "Rock", "Scissors"]
ground = np.argmax(y_valid,axis=1)
print(classification_report(ground,pred,target_names = pred_labels))
```
